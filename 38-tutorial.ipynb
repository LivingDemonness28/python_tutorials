```python
def lexicographical_order():
    # Initialize the list outside the loop
    words_array = []
    
    while True:
        words = input("Please enter a word: ")
        print(" ")
        yn = ""
        while yn not in ["y", "n"]:
            yn = input("Would you like to add another word (y/n): ").strip().lower()
            print(" ")
        words_array.append(words)
        if yn == "n":
            break
    sorted_words = sorted(words_array)
    print("This is the original list of words:")
    print(words_array)
    print(" ")
    print("This is the sorted list of words:")
    print(sorted_words)
```

```python
lexicographical_order()
```
